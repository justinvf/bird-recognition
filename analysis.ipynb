## Split dataset into test and training

```python
fold_builder = FoldBuilder.FoldBuilder()
# Just one split for now, these machines don't have the best performance
# 85% train, 15% test
fold = fold_builder.make_cvd_folds(1, 0.85, 0.15, random_seed=0)[0]
```

#### The fold now has arrays of booleans indicating which clips are testing and which are training

```python
print fold.testing_mask
print fold.training_mask
```

```python
training_clipnames = fold.clipnames[fold.training_mask]
testing_clipnames = fold.clipnames[fold.testing_mask]

print len(training_clipnames)
print len(testing_clipnames)
```

#### Get the MFCC features for each clip

```python
training_mfcc_arrays = [data.get_mfcc_array(filename) for filename in training_clipnames] # N_training_clips x N_mfcc_frames x N_mfcc_coefficients
testing_mfcc_arrays = [data.get_mfcc_array(filename) for filename in testing_clipnames]
```

```python
print testing_mfcc_arrays[0].shape # N_mfcc_frames x N_mfcc_coefficients
```

## Build the Model

#### What window length???

```python
WINDOW_LENGTH = 15 # Edit this
```

#### Window the MFCCs with overlap

```python
window_mfccs = transform.WindowMFCCs(window_length=WINDOW_LENGTH, overlap=0.8)
```

```python
training_windowed_mfccs = window_mfccs.transform(training_mfcc_arrays)
testing_windowed_mfccs = window_mfccs.transform(testing_mfcc_arrays)
```

Each window has a feature array of N_mfcc_fames_per_window x N_mfcc_coefficients

```python
print training_windowed_mfccs[0][0].shape
```

#### How to reduce window dimensionality???

```python
REDUCE_MFCC_ROW = np.max # Edit this
```

#### Reduce the dimensionality of the windows

```python
window_summarizer = transform.SummarizeWindowedMFCCs(REDUCE_MFCC_ROW) # Take the time-average of each mfcc row
```

```python
training_features = window_summarizer.transform(training_windowed_mfccs) # N_training_examples x N_windows x N_features_per_window
testing_features = window_summarizer.transform(testing_windowed_mfccs)
```

```python
print training_features[0].shape # N_windows x N_features_per_window
```

#### Train a random forest by distributing the clip labels to the windows in that clip

```python
clf = classifier.LabelDistributingBinaryForest(
    n_jobs= 1, # In testing this on ec2, n_jobs > 1 was crashing python with a memory error
    n_estimators = 5, # Start low since n_jobs == 1
    min_samples_leaf = 2,
    min_samples_split = 5,
    oob_score=True
)

# Pull out labels for just the training examples
label_matrix = fold.label_df.ix[fold.clipnames[fold.training_mask]].values
clf.fit(training_features, label_matrix)
```

#### Predict the labels for each window

```python
window_predictions = clf.transform(testing_features)
```

#### How to turn window predictions into clip predictions???

```python
COLLAPSE_PREDICTION_ROW = np.min # Edit this
```

#### Convert window predictions to clip predictions

```python
prediction_collapser = transform.CollapseWindowPredictions(COLLAPSE_PREDICTION_ROW)
prediction_matrix = prediction_collapser.transform(window_predictions)
```

```python
print prediction_matrix.shape
print prediction_matrix
```

#### Show the truth

```python
truth_matrix = fold.label_df.ix[fold.testing_mask].values
print truth_matrix.shape
print truth_matrix
```

### Evaluate AUROC metric

```python
sklearn.metrics.auc_score(truth_matrix.ravel(), prediction_matrix.ravel())
```

## Create Kaggle Submission

Takes a while on ec2, especially on just one core

```python
def create_features(list_of_mfcc_arrays):

    window_mfccs = transform.WindowMFCCs(WINDOW_LENGTH, overlap=0.8)
    window_summarizer = transform.SummarizeWindowedMFCCs(REDUCE_MFCC_ROW) # Take the time-average of each mfcc row

    
    windowed_mfccs = window_mfccs.transform(list_of_mfcc_arrays)
    summarized_mfccs = window_summarizer.transform(windowed_mfccs)

    return summarized_mfccs

def fit_classifier(summarized_mfccs, fold):
    
    clf = classifier.LabelDistributingBinaryForest(
        n_jobs= 1,
        # n_estimators = 50,
        n_estimators = 5,
        min_samples_leaf = 2,
        min_samples_split = 5,
        oob_score=False
    )
 
    summarized_mfccs = create_features(list_of_mfcc_arrays)
    clf.fit(summarized_mfccs, fold)
    
    return clf

def window_preds_to_clip_preds(window_predictions):
    
    prediction_collapser = transform.CollapseWindowPredictions(COLLAPSE_PREDICTION_ROW)
    list_of_preds = prediction_collapser.transform(window_predictions)
    return transform.make_prediction_df(list_of_preds, fold)
```

```python
fold_builder = FoldBuilder.FoldBuilder()
fold = fold_builder.submission_fold()
list_of_mfcc_arrays = [data.get_mfcc_array(filename) for filename in fold.clipnames]
```

```python
features = create_features(list_of_mfcc_arrays)
```

```python
clf = fit_classifier(features, fold)
```

```python
window_predictions = clf.transform(features)
```

```python
predictions = window_preds_to_clip_preds(window_predictions)
```

```python
submission_utils.write_submission_csv(predictions, 'submission.csv')
```
